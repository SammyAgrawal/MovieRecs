# User Based Collaborative Filtering

```python
matrix = pd.read_csv('normalized.csv')
matrix.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
from scipy import spatial
```

```python
def cosine_similarity(v1, v2):
    v1 = np.array(v1); v2 = np.array(v2);
    magnitudes = np.sqrt(sum(np.square(v1))) * np.sqrt(sum(np.square(v2)))
    #alternatively
    magnitudes = np.linalg.norm(v1) * np.linalg.norm(v2)
    
    return(sum(v1*v2)/magnitudes)
```

```python
cosine_similarity([1,5], [2,3])
```

```python
def user_similarity(id1, id2):
    return(cosine_similarity(norm.loc[id1], norm.loc[id2]))
```

```python
norm = matrix.fillna(0)
```

```python
predictions = matrix.fillna(0)
```

```python
for user in matrix[6].dropna().index:
    print(user, matrix[6].dropna()[user])
```

```python
def predict(user, movie, k=5):
    other_raters = matrix[movie].dropna() # only look at users that
    similar = []
    for rater in other_raters.index:
        if (rater!=user): #triggered max 1 time
            similar.append((rater, user_similarity(user, rater)))
    top_k = sorted(similar, key=lambda x:x[1], reverse=True)[:k]
    
    r = [(other_raters[i[0]],i[1]) for i in top_k]

    # WEIGHT RESULTS
    
    weighing = 0 # sum of all weights, will divide by 
    total=0
    
    for rating, weight in r:
        weighing+=weight
        total += rating * weight
    return(total/weighing)
    
    
```

```python
for movie in predictions.columns:
    for user in predictions.index:
        predictions[movie][user] = predict(user, movie)
```
